```python
#bullet 1: quadratic
import numpy as np
import matplotlib.pyplot as plt

# This function uses gradient descent to find the lowest point (minimum) of a function
def gradient_descent(f, learning_rate, initial_point):
    # calculates the slope (derivative) of the function at a given point
    def deriv(f, base_point):
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))

    # Lists to keep track of the x values and their corresponding y (function) values
    x_coords = [initial_point]  
    y_coords = [f(initial_point)]  # Calculate the function value at the start

    # Do the gradient descent for 1000 steps
    for _ in range(1000):
        # Find the slope of the function at the current x value
        gradient = deriv(f, x_coords[-1])
        # update x as it moves down the slope
        update = -learning_rate * gradient
        x_coords.append(x_coords[-1] + update)  # add the new x value to the list
        # Calculate and store the function value at the new x value
        y_coords.append(f(x_coords[-1]))

    # prepare points to draw the function curve
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 1000)
    function_range = [f(i) for i in plot_range]

    #plotting
    plt.plot(plot_range, function_range, label='Function f(x)')
    plt.plot(x_coords, y_coords, marker='o', color='r', label='Steps')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.title('Finding the Minimum with Gradient Descent') 
    plt.legend()
    plt.show()

    # Return the final x value (best guess for minimum) and the function value at that x
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# This is the function we are trying to minimize
def f(x):
    return x**2

# set learning rate and the starting point
learning_rate = 0.1
initial_point = 3

# Run the gradient descent function and print the results
optimal_point, optimal_value = gradient_descent(f, learning_rate, initial_point)
print("Optimal Point:", optimal_point)  #this is the x value where the function is smallest
print("Optimal Value:", optimal_value)  #this is the smallest function value
```
